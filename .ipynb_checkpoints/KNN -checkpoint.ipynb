```python
import pandas as pd
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
columns_to_use = ['Product_Category_2','Product_Category_3','Product_Category_1']
```

```python
knn_imputer = KNNImputer(n_neighbors= 5)
df_imputed = df.copy()
```

```python
df_imputed[columns_to_use] = knn_imputer.fit_transform(df[columns_to_use])
```

```python
print('Missing vslues after using KNN')
print(df[columns_to_use].isnull().sum)
```
