```python
## example

# Step 1: Import required libraries
from sklearn.linear_model import LinearRegression

# Step 2: Impute missing values in Product_Category_2 using Product_Category_1

# Split data into rows with and without missing Product_Category_2
df_complete_pc2 = df[df['Product_Category_2'].notnull()]
df_missing_pc2 = df[df['Product_Category_2'].isnull()]

# Independent variable (Product_Category_1)
y_train_pc2 = df_complete_pc2['Product_Category_2']

# Independent variable for the missing data in Product_Category_2
X_missing_pc2 = df_missing_pc2[['Product_Category_1']]

# Step 3: Train the model to predict missing Product_Category_2
reg_pc2 = LinearRegression()
reg_pc2.fit(X_train_pc2, y_train_pc2)

# Step 4: Predict missing values for Product_Category_2
df.loc[df['Product_Category_2'].isnull(), 'Product_Category_2'] = reg_pc2.predict(X_missing_pc2)

# Step 5: Impute missing values in Product_Category_3 using Product_Category_1

# Split data into rows with and without missing Product_Category_3
df_complete_pc3 = df[df['Product_Category_3'].notnull()]
df_missing_pc3 = df[df['Product_Category_3'].isnull()]

# Independent variable (Product_Category_1)
X_train_pc3 = df_complete_pc3[['Product_Category_1']]
y_train_pc3 = df_complete_pc3['Product_Category_3']

# Independent variable for the missing data in Product_Category_3
X_missing_pc3 = df_missing_pc3[['Product_Category_1']]

# Step 6: Train the model to predict missing Product_Category_3
reg_pc3 = LinearRegression()
reg_pc3.fit(X_train_pc3, y_train_pc3)

# Step 7: Predict missing values for Product_Category_3
df.loc[df['Product_Category_3'].isnull(), 'Product_Category_3'] = reg_pc3.predict(X_missing_pc3)

# Step 8: Verify the missing values after regression imputation
print("Missing values after regression imputation:")
print(df.isnull().sum())
```
